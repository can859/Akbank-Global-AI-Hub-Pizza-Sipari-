```python
import csv
from datetime import datetime

menu_list="* Lütfen Bir Pizza Tabanı Seçiniz: 1: Klasik 2: Margarita 3: Türk Pizza 4: Sade Pizza * ve seçeceğiniz malzeme: 11: Zeytin 12: Mantarlar 13: Keçi Peyniri 14: Et 15: Soğan 16: Mısır * Teşekkür ederiz!"


menu_file=open("menu.txt",'w')
menu_file.write(menu_list)
menu_file.close()


class PIZZA:
  def __init__(self,description,cost):
    self.description=description
    self.cost=cost

  def get_description(self):
    return self.description

  def get_cost(self):
    return self.cost

class PIZZA_BASE(PIZZA):
  def __init__(self,description,cost):
    self.description=description
    self.cost=cost

  def get_description(self):
    return self.description

  def get_cost(self):
    return self.cost

class KLASIK(PIZZA_BASE):
  def __init__(self):
    self.description='Klasik Pizza'
    self.cost=125

class MARGARITA(PIZZA_BASE):
  def __init__(self):
    self.description='Margarita Pizza'
    self.cost=100

class TURKPIZZA(PIZZA_BASE):
  def __init__(self):
    self.description='Türk Pizza'
    self.cost=130

class SADEPIZZA(PIZZA_BASE):
  def __init__(self):
    self.description='Sade Pizza'
    self.cost=95

class DECORATOR(PIZZA):
  def __init__(self,pizza):
    self.pizza=pizza

  def get_description(self):
    return self.pizza.get_description() + ' - ' + self.description

  def get_cost(self):
    return self.pizza.get_cost() + self.cost

class ZEYTIN(DECORATOR):
  def __init__(self,pizza):
    self.pizza=pizza
    self.description='Zeytin'
    self.cost=10

class MANTAR(DECORATOR):
  def __init__(self,pizza):
    self.pizza=pizza
    self.description='Mantar'
    self.cost=8

class KECIPEYNIRI(DECORATOR):
  def __init__(self,pizza):
    self.pizza=pizza
    self.description='Keçi Peyniri'
    self.cost=10

class ET(DECORATOR):
  def __init__(self,pizza):
    self.pizza=pizza
    self.description='Et'
    self.cost=15

class SOGAN(DECORATOR):
  def __init__(self,pizza):
    self.pizza=pizza
    self.description='Soğan'
    self.cost=5

class MISIR(DECORATOR):
  def __init__(self,pizza):
    self.pizza=pizza
    self.description='Mısır'
    self.cost=7

def main():
  print("Akbank Pizza'ya Hoş Geldiniz!")
  print("Menümüz:")
  with open("menu.txt", 'r') as f:
    menu=f.read()
    print(menu)

    taban_secimi= None
    while not taban_secimi:
      try:
        taban_secimi=int(input("1-4 pizza tabanları arasından bir seçim yapınız: "))
        if taban_secimi not in [1,2,3,4]:
          raise ValueError
      except:
        print("Lütfen menüdeki tabanlardan seçiniz!")
        taban_secimi=None
            
    
    if taban_secimi==1:
      pizza=KLASIK()
    elif taban_secimi==2:
      pizza=MARGARITA()
    elif taban_secimi==3:
      pizza=TURKPIZZA()
    else:
      pizza=SADEPIZZA()

    print("Taban seçiminiz: " + str(pizza.get_description()))
    

    while True:
      malzeme_secimi=int(input("11-16 arasından sadece bir malzeme seçiniz ya da işleminiz bittiyse '20' yazınız: "))

      if malzeme_secimi==11:
        pizza=ZEYTIN(pizza)
      elif malzeme_secimi==12:
        pizza=MANTAR(pizza)
      elif malzeme_secimi==13:
        pizza=KECIPEYNIRI(pizza)
      elif malzeme_secimi==14:
        pizza=ET(pizza)
      elif malzeme_secimi==15:
        pizza=SOGAN(pizza)
      elif malzeme_secimi==16:
        pizza=MISIR(pizza)
      elif malzeme_secimi==20:
        break
      else:
        print("Lütfen menüden bir malzeme seçiniz!")
        
    
    print("Checkout: " + str(pizza.get_description()))

    total_cost=pizza.get_cost()
    print("Borcunuz: " + str(total_cost)+ " TL'dir")

    isim= input("İsminizi giriniz. ")
    while True:
      try:
        tc_no=int(input("TC numaranızı giriniz. "))
        if len(str(tc_no))!=11:
          print("11 haneli TC numaranızı doğru giriniz!")
        else:
          break
      except ValueError:
        print("Sayı değeri girmelisiniz!")
    while True:
      try:
        kredi_karti_no=int(input("Kredi kartı numaranızı giriniz. "))
        if len(str(kredi_karti_no))!=16:
          print("16 haneli kredi kartı numaranızı doğru giriniz!")
        else:
          break
      except ValueError:
        print("Sayı değeri girmelisiniz!")
    while True:
      try:
        kredi_karti_sifre=int(input("Kredi kartı şifrenizi giriniz. "))
        if len(str(kredi_karti_sifre))!=4:
          print("4 haneli kredi kartı şifrenizi doğru giriniz!")
        else:
          break
      except ValueError:
        print("Sayı değeri girmelisiniz!")

    with open('orders_database.csv', 'w', newline='') as file:
      writer=csv.writer(file)
      writer.writerow([taban_secimi, malzeme_secimi, PIZZA.get_description, PIZZA.get_cost, isim, tc_no, kredi_karti_no, kredi_karti_sifre, datetime.now()])


if __name__ == "__main__":
    main()
```
